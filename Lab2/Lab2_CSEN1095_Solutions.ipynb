## Bar Chart Exercise

```python
#Read the CSV
df_fifa = pd.read_csv('https://raw.githubusercontent.com/raneemsultan/DataEngineering-CS1095/main/Lab2/CSVs/fifa.csv')
#Observe the rows
df_fifa.head()
```

```python
#Bar Chart
nationality_AVGoverall_fifa_series= df_fifa.groupby('Nationality')['Overall'].mean() #get the mean overall rating based on nationality
nationality_AVGoverall_fifa_series =nationality_AVGoverall_fifa_series.sort_values(ascending=False)#sort the results descendingly 
nationalities = nationality_AVGoverall_fifa_series[:10].index.tolist()#get the nationalities of the top 10 results
nationalities_average_overall = nationality_AVGoverall_fifa_series[:10].values.tolist() #get the overall average rating of the top 10 results
print(nationality_AVGoverall_fifa_series[:10])#print the top 10 countries and their corresponding average rating

figFIFA, axFIFA = plt.subplots(figsize=(8,4),dpi=100) #set the size of the figure
SeabornBARCHART = sns.barplot(x=nationalities,y=nationalities_average_overall,color='black')#plot the barchart
SeabornBARCHART.set(xlabel="Country of Origin", ylabel="Average Overall Rating",title="Average Overall Rating VS Nationality")#set the labels and title of the graph
SeabornBARCHART.set_xticklabels(SeabornBARCHART.get_xticklabels(), rotation=70,fontdict={'size':10})#style the x ticks
```

## Advanced Bar Chart Excerise

**There are several correct/ acceptable / easier approaches to solving this questions. One of the answers proposed by a student is listed below:**

```python
#Read the CSV
df_fifa = pd.read_csv('https://raw.githubusercontent.com/raneemsultan/DataEngineering-CS1095/main/Lab2/CSVs/fifa.csv')
#Observe the rows
df_fifa.head()
#==========================================================================================
#Filter and aggregate
nationalityAggregated = df_fifa.groupby("Nationality")["Overall"].agg(['mean', 'count']) #get the mean and count for every nationality
nationalityAggregated = nationalityAggregated[nationalityAggregated['count']>30]['mean'].sort_values(ascending=False)
nationalityAggregated = nationalityAggregated[:10]
#==========================================================================================
#plot the barchart
figFIFA, axFIFA = plt.subplots(figsize=(8,4),dpi=100) #set the size of the figure
SeabornBARCHART = sns.barplot(x=nationalityAggregated.index.values.tolist(),y=nationalityAggregated.values.tolist(),color='black')#plot the barchart
SeabornBARCHART.set(xlabel="Country of Origin", ylabel="Average Overall Rating",title="Average Overall Rating VS Nationality")#set the labels and title of the graph
SeabornBARCHART.set_xticklabels(SeabornBARCHART.get_xticklabels(), rotation=70,fontdict={'size':10})#style the x ticks
```

**Another solution would be:**

```python
#Read the CSV
df_fifa = pd.read_csv('https://raw.githubusercontent.com/raneemsultan/DataEngineering-CS1095/main/Lab2/CSVs/fifa.csv')
#Observe the rows
df_fifa.head()

#Bar Chart
nationalityCol = df_fifa.groupby("Nationality")["Nationality"].transform(len)#transform creates a new col that will hold the number of times a certain nationality occured, and will place that number in each row. 
nationality_mask = (nationalityCol > 30) #create a mask that gets us the rows with nationalityCol> 10 only
df_fifa =df_fifa[nationality_mask]#mask the dataframe and set the dataframe to the new masked one

nationality_AVGoverall_fifa_series= df_fifa.groupby('Nationality')['Overall'].mean() #get the mean overall rating based on nationality
nationality_AVGoverall_fifa_series =nationality_AVGoverall_fifa_series.sort_values(ascending=False)#sort the results descendingly 
nationalities = nationality_AVGoverall_fifa_series[:10].index.tolist()#get the nationalities of the top 10 results
nationalities_average_overall = nationality_AVGoverall_fifa_series[:10].values.tolist() #get the overall average rating of the top 10 results
print(nationality_AVGoverall_fifa_series[:10])#print the top 10 countries and their corresponding average rating

figFIFA, axFIFA = plt.subplots(figsize=(8,4),dpi=100) #set the size of the figure
SeabornBARCHART = sns.barplot(x=nationalities,y=nationalities_average_overall,color='black')#plot the barchart
SeabornBARCHART.set(xlabel="Country of Origin", ylabel="Average Overall Rating",title="Average Overall Rating VS Nationality")#set the labels and title of the graph
SeabornBARCHART.set_xticklabels(SeabornBARCHART.get_xticklabels(), rotation=70,fontdict={'size':10})#style the x ticks
```

## HeatMap Excerise 

```python
# Read the data
df_gas = pd.read_csv("https://raw.githubusercontent.com/raneemsultan/DataEngineering-CS1095/main/Lab2/CSVs/gas_prices.csv",index_col=0)
# Set the width and height of the figure
plt.figure(figsize=(14,7))
# Heatmap showing the gas price for each country over the years
gas_hm= sns.heatmap(data=df_gas.loc[2000:], annot=True)
gas_hm.set(xlabel="Country", ylabel="Year",title="Gas Price")#set the labels and title of the graph
```
